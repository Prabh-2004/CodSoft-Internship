# Data Preprocessing

```python
df.head()
```

## Defining Inputs and the Target

```python
# defining the inputs and the target
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
X.head()
```

```python
y
```

## Train test split

```python
# splitting the inputs and the target into training and testing data.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

## Standardization

```python
# Standardizing the inputs and the target ->
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler()

X_train = X_scaler.fit_transform(X_train)
```

# Model Training and Testing

```python
# Accuracy Method.
from sklearn.metrics import r2_score, mean_squared_error
def accuracy(X, y, model):
    '''This Method calculates r_squared and mean_squared_error values based on 
    given inputs and target.'''
    y_pred = model.predict(X)
    rmse = np.sqrt(mean_squared_error(y_pred, y))
    r_squared = r2_score(y_pred, y)
    
    print("The R_squared value: ", r_squared)
    print("The Root Mean squared error: ", rmse)
```

```python
# K-Fold Cross Validation.
from sklearn.model_selection import cross_val_score
def k_fold_cross_val(model, X, y):
    '''This Method is used to calculate the mean accuracy and the std.dev. using 
    k Fold cross validation.'''
    accuracies = cross_val_score(estimator=model, X=X, y=y, cv=10)
    print("Mean RÂ² Score: ", (accuracies.mean()*100).round(2), "%")
    print("Standard Deviation: ", (accuracies.std()*100).round(2), "%")
```

## Linear Regression

```python
# Multiple Linear Regression -> 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Training accuracy -> 
accuracy(X_train, y_train, lr)
```

```python
# Test Accuracy ->
X_test = X_scaler.fit_transform(X_test)

accuracy(X_test, y_test, lr)
```

```python
# K-Fold Cross Validation of ten random samples ->
k_fold_cross_val(lr, X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=3)
X_train_poly = poly_features.fit_transform(X_train)

pr = LinearRegression()
pr.fit(X_train_poly, y_train)
```

```python
# Training accuracy -> 
accuracy(X_train_poly, y_train, pr)
```

```python
# Test Accuracy ->
X_test = X_scaler.fit_transform(X_test)
X_test_poly = poly_features.fit_transform(X_test)

accuracy(X_test_poly, y_test, pr)
```

```python
# K-Fold Cross Validation of ten random samples ->
k_fold_cross_val(pr, X_train_poly, y_train)
```

## Support Vector Regressor

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
```

```python
# Training accuracy -> 
accuracy(X_train, y_train, svr)
```

```python
# Test Accuracy ->
X_test = X_scaler.fit_transform(X_test)

accuracy(X_test, y_test, svr)
```

```python
# K-Fold Cross Validation of ten random samples ->
k_fold_cross_val(svr, X_train, y_train)
```

## Decision tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=69)
dtr.fit(X_train, y_train)
```

```python
# Training accuracy -> 
accuracy(X_train, y_train, dtr)
```

```python
# Test Accuracy ->
X_test = X_scaler.fit_transform(X_test)

accuracy(X_test, y_test, dtr)
```

```python
# K-Fold Cross Validation of ten random samples ->
k_fold_cross_val(dtr, X_train, y_train)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=10, random_state=0)
rfr.fit(X_train, y_train)
```

```python
# Training accuracy -> 
accuracy(X_train, y_train, rfr)
```

```python
# Test Accuracy ->
X_test = X_scaler.fit_transform(X_test)

accuracy(X_test, y_test, rfr)
```

```python
# K-Fold Cross Validation of ten random samples ->
k_fold_cross_val(rfr, X_train, y_train)
```
